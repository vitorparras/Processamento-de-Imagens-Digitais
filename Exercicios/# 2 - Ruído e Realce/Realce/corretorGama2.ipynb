```python
# Usei uma função para calcular a correção gama
# Depois utilizei um laço 'for' para percorrer os valores dado pelo professor e trazer as imagens
# Em relação as métricas não tenho certeza de estarem corretas, Jaccard e erro médio quadrático
# A imagem que usei foi do exercício 6 e tinha realizado uma transformação nem em gaussian




import cv2
import numpy as np
from matplotlib import pyplot as plt

def gama(img, c=1):
   
    invGama = 1 / c
    correcao_gama = np.array([((i / 255) ** invGama) * 255 for i in np.arange(0, 256)]).astype("uint8")
    
    return cv2.LUT(imagem, correcao_gama)

imagem = cv2.imread('egaussian.jpg', 1)


def jaccard_binary(x, y):
    intersection = np.logical_and(x, y)
    union = np.logical_or(x, y)
    similiaridade = intersection.sum() / float(union.sum())
    
  
    return similiaridade



for i in [0.04, 0.4, 2.5, 10]:

    plt.axis("off")
    imagem1 = gama(imagem, i)
    plt
    img = np.hstack((imagem, imagem1))
    plt.title("       Original                     Correção Gama")
    similarity = jaccard_binary(imagem, imagem1)
    emq = np.square(np.subtract(imagem, imagem1)).mean()
    plt.imshow(cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE))
    plt.show()
    print('Coeficiente de jaccard = ', similarity)
    print('Erro médio quadrático = ', emq)


```
